### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is lightweight and flexible python web framework that is widely used for developing web applications. 
Flask is known for its simplicity, ease of use, and flexibility, which makes it a popular choice among developers for building web applications.

The advantages of Flask Framework:

    1. Easy to learn:
            Flask has a simple and intuitive API, which makes it easy for beginners to learn and use.
    2. Lightweight:
            Flask is a lightweight framework and does not require any particular tools or libraries to get started.
    3. Flexible:
            Flask is flexible and can be easily customized to suit the specific requirements of the project.
    4. Modular:
            Flask is built with a modular design, which makes it easy to add or remove features as per the project needs.
    5. Supports multiple databases:
            Flask support multiple databases, including MySQL, PostgreSQL, SQLite and others.
    6. Extensible:
            Flask is extensible, which means that you can use various Flask extensions to add extra functionality to your application.

### Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping URLs to specific functions in your Flask application. Flask provides a built-in decorator called @app.route() that allows you to define routes for your application.

We use app routes in Flask to define the URLs that users can visit to access different parts of our application. By defining routes, we can separate our application into different views, making it easier to manage and maintain.

### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
    route to show the following details:
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    
    Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def home():
    return '''Company Name: ABC Corporation</br>
                Location: India</br>
                Contact Detail: 999-999-9999'''

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. This function generates a URL for a given view function name and any arguments that the view function requires.

```python
from flask import Flask,url_for

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return '<h1>Welcome to ABC Corporation</h1>'

@app.route('/')
def home():
    return '''Company Name: ABC Corporation</br>
                Location: India</br>
                Contact Detail: 999-999-9999'''

if __name__=="__main__":
    with app.test_request_context():
        print(url_for('welcome'))
        print(url_for('home'))
```

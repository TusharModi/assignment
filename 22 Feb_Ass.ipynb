Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

```python
import numpy as np
data = {
'name': ['Alice', 'Bob', 'Charlie', 'David', np.nan, 'Frank', 'Grace', 'Hannah', np.nan,np.nan,'Alice'],
'age': [25, 36, 28, 42, 29, 31, 27, 24, 35, 33,25],
'gender': ['female', 'male', 'male', 'male', 'female', 'male', 'female', 'female', 'male', 'male','female'],
'country': ['USA', 'Canada', 'UK', 'USA', 'USA', 'Canada', 'UK', np.nan, 'Canada', np.nan,'USA'],
'income': [45000, 55000, 60000, 70000, 35000, 80000, 50000, 42000, 75000, 62000,45000]
}
df=pd.DataFrame(data)
df
```

### 1. Value_counts

```python
df.value_counts()
```

### 2. Sort_index

```python
df.sort_index()
```

### 3. Total duplicated value

```python
df.duplicated().sum()
```

### 4. Unique

```python
df['name'].unique()
```

### 5. Total nunique number

```python
df['name'].nunique()
```

### 6. drop_duplicates

```python
df.drop_duplicates()
```

### 7. Fill the blank data.

```python
df.fillna(method='ffill')
```

### 8. Set Index

```python
df.set_index('name')
```

### 9. reset Index change index

```python
df.reset_index().set_index('gender')
```

### 10. Rank

```python
df['rank_name']=df['income'].rank(ascending=False)
df.sort_values('rank_name')
```

Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
courses = pd.Series(["Spark","PySpark","Hadoop"],name='Courses')
fees = pd.Series([22000,25000,23000],name='fees')
discount  = pd.Series([1000,2300,1000],name='discounts')
```

```python
df=pd.concat([courses,fees],axis=1)
df
```

```python
df=pd.concat([courses,fees,discount],axis=1)
df
```
